```python

def generate_summary(base_dir,ftype):
    
    df = pd.concat(pd.read_csv(base_dir + '/' + f,encoding= 'unicode_escape') for f in os.listdir(base_dir))
    
    df['burndown_flag'] = ftype
    
    print(df.groupby('Date')['Device Name'].agg({'count'}).reset_index())
            
    return df
```

```python
raw_base_dir = '/Users/kaipak/Desktop/cyber-vulnerability/data/itsec-edw/cleaned_files/HIGH_CRITICAL/historical/raw'

my_sum = generate_summary(raw_base_dir, 'raw')
```

```python
SEARCH_TERM = "NISAR.*GDS"
summary = ['Date', 'Security Plan', 'Security Plan Name', 'Is Mission']


summary = (
    my_sum.loc[my_sum['Security Plan Name'].str.contains(SEARCH_TERM, na=False)][summary].
    groupby(by=summary).
    size().
    to_frame('instances').
    reset_index().
    sort_values(by='Date')
)
```

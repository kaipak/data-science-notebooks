```python
def correct_sp_names(df,unique_sp_lookup_path='/Users/kaipak/Desktop/cyber-vulnerability/data/itsec-edw/raw_files/lookups/uq_sp_lookup_cleaned.xlsx'):
    
    print(f"Shape of dataframe before SP Name Correction: {df.shape}")
    unique_sp_lookup = pd.read_excel(unique_sp_lookup_path,converters={'unique_sp_code':str})
    
    # Cleave off last four digits of Security Plan Code. This should be the unique identifier over time
    df['unique_sp_code'] = df['Security Plan'].str.split("-",expand=True)[4]
    
    # Merge data with lookup
    df = df.merge(unique_sp_lookup,on='unique_sp_code',how='left')
    
    # If there is not a Security Plan Code listed in the lookup file, use the one present in original df
    df['Security Plan (Lookup)'] = df['Security Plan (Lookup)'].fillna(value=df['Security Plan'])
    
    # If there is not a Security Plan Name listed in the lookup file, use the one present in original df
    # But favor lookup value since this is ground truth.
    df['Security Plan Name'] = df['Security Plan Name (Adjusted)'].fillna(value=df['Security Plan Name'])
    df['Security Plan Name (Adjusted)'] = df['Security Plan Name (Adjusted)'].fillna(value=df['Security Plan Name'])
    # If this results in a valid security plan name. Backfill the 'Security Plan Name' field so joins will work correctly. 
    # DEV
    #df.loc[pd.isna(df['Security Plan Name']), 'Security Plan Name'] = df['Security Plan Name (Adjusted)']
    #df.loc[(df['Security Plan Name'] == 'Unknown'), 'Security Plan Name'] = df['Security Plan Name (Adjusted)']
    
    # Append Security Plan Code to Security Plan Name for clarify
    df['Security Plan Name (Adjusted)'] = df['Security Plan Name (Adjusted)'] + ' (' + df['Security Plan (Lookup)'] + ')'
    
    print(f"Shape of dataframe after SP Name Correction: {df.shape}")
    return df
```
